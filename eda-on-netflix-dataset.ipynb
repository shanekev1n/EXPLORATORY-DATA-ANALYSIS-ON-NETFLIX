**Below libraries are used for EDA analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Next we are going to read the netflix dataset in the form csv format**

```python
data = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
data
```

**Before going to further Analysing,first we need to if there is any null values**

```python
data.isnull().sum() #checking the number of null values in the dataset
```

**The next step is to treat the null values with appropriate values**

```python
data['director'] = data['director'].fillna('No Data Availabe')
data['cast'] = data['cast'].fillna('No Data Available')
data['country'] = data['country'].fillna('No Data Available')
data['date_added'] = data['date_added'].fillna(data['date_added'].mode()[0])  
data['rating'] = data['rating'].fillna(data['rating'].mode()[0])
data['duration'] = data['duration'].fillna('No Data Available')
```

**Now check whether there is any null values in the dataset**

```python
data.isnull().sum() 
```

# Visualization

**Which is the most watched show on Netflix?**

```python
TV_shows =data.loc[data['type'] == "TV Show"].count()[0]
movies = data.loc[data['type'] == 'Movie'].count()[0]
                                                   
#assigning colors,labels amd title to the chart
colors=('red','blue')
labels =['TV Show','Movie']
plt.title("Netflix Content in %",fontdict = {'fontweight':'bold','fontsize':18})

#creating a pie chart

piechart = plt.pie([TV_shows,movies],colors=colors,labels=labels,autopct ='%.2f %%')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

**Insight:**
* **Probably 7  out 10 peoples prefer to watch Movies rather than TV Show**

**Which are the top 10 countries with most contribution to Netflix**

```python
#creating a bar chart
data.country.value_counts()[:10].plot(kind="bar")
plt.title('Top 10 countries with Most Contributions to Netflix',fontdict={"fontweight":"bold"})
plt.xlabel('Country')
plt.ylabel('')
plt.show()
```

**Insights:**
* **From stats it is  evident that Inited states is the country with most contribution to netflix followed by India the second most contributed country to netflix**

**Who are the top 10 directors with most contributions to Netflix?**

```python
#creating a bar chart
data.director.value_counts()[1:11].plot(kind="bar")
plt.yticks(np.arange(0,22,1))
plt.title("Top 10 directors with Most Netflix Shows",fontdict={'fontweight':'bold'})
plt.xlabel("Directors")
plt.ylabel("Total Shows")
plt.show()
```

**Insights:**
* **Rajiv Chilaka is the most contributed director with 19 contributions to the netflix**

**Which is having the top 3 highest rating?**

```python
#creating a bar chart
x = list(data.rating.value_counts().index)
y=list(data.rating.value_counts().values)
plt.subplots(figsize =(16, 9))
plt.bar(x,y,color ='maroon')
#displaying values on top of the bar
for i, v in enumerate(y):
    plt.text(i,v+60,str(v),ha ="center",weight="bold",size=10)
plt.title('Distribution of Ratings')
plt.show()
```

**Insights:**
* **TV-MA has the highest number of rating(i.e,3211) followed by TV-14(2160) and TV-PG(863)**
* **74 min,84 min and 66 min are equally**
